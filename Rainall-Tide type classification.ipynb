```python
# 计算每场事件的降雨强度
# 计算每场事件的所有单元总降雨量

# 计算每场事件的总降雨时长
# 计算每场事件的平均降雨强度

results = []

for df in dfn_s:
    # 对RH列求和
    rh_sum = df['RH'].sum()
    # print(rh_sum)
    # 计算FID_为0的个数
    fid_0_count = (df['FID_'] == 0).sum()
    # print(fid_0_count)
    fid_0_count_min = fid_0_count * 60
    # print(fid_0_count_min)
    # 计算结果并添加到结果列表中，保留两位小数
    result = (rh_sum / fid_0_count_min).round(2)
    results.append(result)

# 使用sorted函数对结果进行排序（从大到小）
sorted_results = sorted(enumerate(results), key=lambda x: x[1], reverse=True)

# 打印排序后的结果
for i, result in sorted_results:
    print(result)
    # print(f"dfn_{i + 1}的平均降雨强度为: {result}")
# 将结果输出为CSV文件
df_results = pd.DataFrame(sorted_results, columns=['event_id', 'average_rainfall_intensity'])
df_results.to_csv("D:/a. work/papers/interpretable ML for urban flood susceptibility/数据类型对比分析/average_rainfall_intensity.csv", index=False)
```
